```python
def count_words(text):
    word_counts = {}
    skip = [".",",",";",":","'",'"']
    for ch in skip:
        text = text.replace(ch,"")
    for word in text.lower().split(" "):
        if word in word_counts.keys():
            word_counts[word] += 1
        else:
            word_counts[word] = 1
    return word_counts
```

```python
from collections import Counter
```

```python
def count_words_fast(text):
    word_counts = {}
    skip = [".",",",";",":","'",'"']
    for ch in skip:
        text = text.replace(ch,"")
    text = text.lower()
    
    word_counts = Counter(text.split())
    return word_counts
```

```python
count_words_fast(text)
```

```python
def count_words(text):
    text = text.lower()
    skips = ['.', ',',':', ';', "'", '"']
    for ch in skips:
        text = text.replace(ch, "")
    
    word_counts = {}
    for word in text.split(" "):
        if word in word_counts:
            word_counts[word] +=1 
        else:
            word_counts[word] = 1
    return word_counts
```

```python
def read_book(title_path):
    with open(title_path, 'r', encoding = 'UTF-8') as current_file:
        text = current_file.read()
        text = text.replace('\n', '').replace('\r', '')
    return text
```

```python
def word_stats(word_counts):
    num_unique = len(word_counts)
    counts = word_counts.values()
    return (num_unique, counts)
```

```python
import os
book_dir = './books'

import pandas as pd
stats = pd.DataFrame(columns = ('Language', 'Author', 'Title', 'Lenght', 'Unique Words' ))
title_num = 1
```

```python
for language in os.listdir(book_dir):
    if language == ".DS_Store":
        continue
    for author in os.listdir(book_dir + "/" + language):
        if author == ".DS_Store":
            continue
        for title in os.listdir(book_dir + "/" + language + '/' + author):
            if title == ".DS_Store":
                continue
            inputfile = book_dir + '/' + language + '/' + author + '/' + title
            print(inputfile)
            text = read_book(inputfile)
            (num_unique, counts) = word_stats(count_words(text))
            stats.loc[title_num] = language, author, title, sum(counts), num_unique
            title_num += 1
            
```
