```python
import numpy as np
import scipy.stats as ss
import matplotlib.pyplot as plt
n = 100
beta_0 = 5
beta_1 = 2
np.random.seed(1)
x = 10 * ss.uniform.rvs(size=n)
y = beta_0 + beta_1 * x + ss.norm.rvs(loc=0, scale = 1, size = n)
```

```python
rss = []
slopes = np.arange(-10,15,0.01)
```

```python
for slope in slopes:
    rss.append(np.sum((y - beta_0 - slope * x)**2))
```

```python
ind_min = np.argmin(rss)
```

```python
slopes[ind_min]
```

```python
rss[ind_min]
```

```python
plt.figure()
plt.plot(slopes,rss)
plt.plot(1.9999999999997442, 81.54000742551887, "*")
plt.xlabel('Slope')
plt.ylabel('RSS')
plt.show()
```
