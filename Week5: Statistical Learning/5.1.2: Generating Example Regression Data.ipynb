```python
import numpy as np
import scipy.stats as ss
import matplotlib.pyplot as plt
```

```python
n = 100
beta_0 = 5
beta_1 = 2
np.random.seed(1)
```

```python
x = 10 * ss.uniform.rvs(size=n)
y = beta_0 + beta_1 * x + ss.norm.rvs(loc=0, scale = 1, size = n)
```

```python
plt.figure()
plt.plot(x,y,'*', ms=5)
xx = np.array([0,10])
plt.xlabel('x')
plt.ylabel('y')
plt.plot(xx, beta_0 + beta_1 * xx)
```
